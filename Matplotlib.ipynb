```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(50)*100
y_data = np.random.random(50)*100
```

# **1. scatter plot**

```python
#                                            s = size of marker  alpha = اذا الراتا كبيرة وفي تداخل بينهن
plt.scatter(X_data, y_data, c="red", marker="*", s = 20)
plt.show()
```

```python
X_data2 = np.random.random(1000)*100
y_data2 = np.random.random(1000)*100

#                                s = size of marker  alpha = اذا الراتا كبيرة وفي تداخل بينهن
plt.scatter(X_data2, y_data2,  s = 20, alpha=0.3)
plt.show()
```

# **2. line chart**

```python
# تغير الوزن خلال السنوات
years = [2006 + x for x in range(16)]
weights = [80, 83, 85, 86, 82, 80, 78, 75,
           77, 81, 88, 92, 85, 80, 79, 77 ]
```

```python
'''
lw = line width

'''
plt.plot(years, weights, c = "g", lw = 3, linestyle = "--") # == (years, weights, "g--", lw = 3)
plt.show()
```

# **3. Bar Plots**
فرقو عن الهيستوغرام أنو هاد للتصنيف و المجموعات\
الهيستوغرام لعدد التكرارات

```python
# عدد الأشخاص المستخدمين
x = ["c++", "c#", "python", "Java", "R"]
y = [50, 60, 150, 20, 70]

plt.bar(x, y, color = "#AC1278", width = 0.5, edgecolor = "#7800AB", lw = 4)
plt.show()
```

# **4. Histograms**

```python
#                       mean, std, num
ages = np.random.normal(20, 1.5, 1000)

# bins = عدد التقسيمات أو نحدد المجالات
# plt.hist(ages,bins=[ages.min(), 18, 21, ages.max()] ,lw = 3)
plt.hist(ages, bins=20, cumulative=True)
```

# **5. Pie chart**

```python
x = ["c++", "c#", "python", "Java", "R"]
y = [50, 60, 150, 20, 70]
plt.pie(y, labels = x)
```

```python
x = ["c++", "c#", "python", "Java", "R"]
y = [50, 60, 150, 20, 70]
explodes = [0.2, 0, 0, 0, 0]
plt.pie(y, labels = x, explode = explodes)
```

```python
x = ["c++", "c#", "python", "Java", "R"]
y = [50, 60, 150, 20, 70]

'''
pctdistance = بعد كتابة النسبة عن المركز

startangle = مكان بداية الرسم

'''
plt.pie(y, labels = x, autopct = "%.2f%%",pctdistance = 1.5, startangle = 90 )
```

# Legends & Multiple Plots

```python
stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [98, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

#        x = generate auto ,y =
plt.plot(stock_a, label = "Company 1")
plt.plot(stock_b, label = "Company 2")
plt.plot(stock_c, label = "Company 3")

# لعرض الأسماء لكل خط
plt.legend(loc = "lower center")
```

```python
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]

plt.pie(votes, labels=None)
plt.legend(labels = people)
```

# Multiple Figures

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("cosine Wave")

axs[1, 0].plot(x, np.log(x))
axs[1, 0].set_title("Log Wave")

axs[1, 1].plot(x, np.random.random(100))
axs[1, 1].set_title("random function")
```

#  Exporting Plots

```python
x2, y2 = np.arange(100), np.random.random(100)

plt.plot(x2, y2)

plt.show()
plt.savefig("y.jpg", dpi = 300)
```

# **3D Plotting**
